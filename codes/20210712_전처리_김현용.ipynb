```python
# 전통시장에 해당하는 row만 추출
is_price_trd = price_all['시장유형 구분(시장/마트) 이름'] == '전통시장'
price_trd = price_all[is_price_trd]
price_trd
```

```python
# 분석에 필요한 열만 추출 및 생성
df_price = price_trd[['년도-월','시장/마트 이름','품목 이름', '실판매규격', '가격(원)', '자치구 이름']]
df_price.rename(columns = {'시장/마트 이름':'시장이름', '품목 이름':'품목이름', '자치구 이름':'자치구명'}, inplace = True)
df_price
```

```python
# 데이터 타입 확인
print(' 년도-월 : ', type(df_price['년도-월'][0]),'\n', '가격(원) : ', type(df_price['가격(원)'][0]))
```

```python
# 유동인구 데이터와 호환을 위한 년도, 분기 column 생성
year = []
quarter = []
for i in range(len(df_price['년도-월'])):
    if df_price['년도-월'][i][5:7] in ['01','02','03'] :
        quarter.append('1')
    if df_price['년도-월'][i][5:7] in ['04','05','06'] :
        quarter.append('2')
    if df_price['년도-월'][i][5:7] in ['07','08','09'] :
        quarter.append('3')
    if df_price['년도-월'][i][5:7] in ['10','11','12'] :
        quarter.append('4') 
    year.append(df_price['년도-월'][i][0:4])
```

```python
# 유동인구 테이블의 년도, 분기 모두 int이므로 str -> int로 변환
for i in range(len(year)):
    year[i] = int(year[i])
    
for i in range(len(quarter)):
    quarter[i] = int(quarter[i])  
    
print(type(year[0]),type(year[len(year)-1]),type(quarter[0]),type(quarter[len(quarter)-1]))
```

```python
# int로 생성한 년도, 분기 column 삽입
df_price.insert(0, '년도', year, True)
df_price.insert(1, '분기', quarter, True)
df_price
```

#### 가격테이블 - 주소테이블 시장이름 차이 조정

- 방배종합시장 -> 남부종합시장  
- 신창시장 -> 창동신창시장  
- 남성시장 -> 사당시장  
- 송화시장 -> 송화골목시장  
- 영동시장 -> 영동전통시장  



- 남구로시장 <- 남구로시장(구 구로자율시장)
- 공릉동도깨비시장 <-공릉동 도깨비시장 

도곡시장 -> 주소테이블에 누락  
방림시장 -> 20년도부터 빠져있음. 시장폐쇄가능성 높음  

```python
# 가격테이블에서 시장이름 주소테이블과 동일하게 변경

def change_name1(x):
    if x=='방배종합시장':
        return "남부종합시장"
    if x=='신창시장':
        return "창동신창시장"
    if x=='남성시장':
        return "사당시장"
    if x=='송화시장':
        return "송화골목시장"
    if x=='영동시장':
        return "영동전통시장"
    if x=='공릉동 도깨비시장':
        return "공릉동도깨비시장"
    else: return x
    
df_price['시장이름'] = df_price['시장이름'].apply(change_name1)
```

```python
def change_name2(x):
    if x=='공릉동 도깨비시장':
        return "공릉동도깨비시장"
    if x=='남구로시장(구 구로자율시장)':
        return "남구로시장"
    else: return x
df_address = df_address.copy()    
df_address['시장이름'] = df_address['시장이름'].apply(change_name2)
```

```python
# 도곡시장, 방림시장 row 삭제

is_true = df_price[df_price['시장이름']=='도곡시장'].index
df_price.drop(is_true, inplace=True)
is_true = df_price[df_price['시장이름']=='방림시장'].index
df_price.drop(is_true, inplace=True)
```

```python
#가격테이블과 주소테이블 merge
df_price_address = pd.merge(df_price, df_address, on='시장이름', how='left')
df_price_address
```

```python
# 가격-주소 테이블에 있는 전통시장 수 확인
print(len(df_price_address['시장이름'].unique()), df_price_address['시장이름'].unique())
```

```python
# 가격-주소 테이블과 유동인구 테이블의 겹치는 시장이름 확인
price_address_market_name = list(df_price_address['시장이름'].unique())
customer_market_name = list(df_customer['시장이름'].unique())

market_name = []
for i in range(len(price_address_market_name)):
    for j in range(len(customer_market_name)):
        if price_address_market_name[i] == customer_market_name[j] :
            market_name.append(price_address_market_name[i])
print(len(market_name), market_name)       
```

#### 가격-주소테이블 - 유동인구테이블 시장이름 차이 조정

- 장위골목시장 -> 장위전통시장  
- 용문시장 -> 용문전통시장  
- 신원시장(신림1동) -> 신원시장  
- 수유재래시장 -> 수유전통시장  
- 원당종합시장 -> 인헌시장  
- 마천중앙시장 -> 마천시장  
- 관악신사시장(신림4동) -> 관악신사시장  



- 대림시장 <- 대림골목시장

```python
# 가격-주소테이블에서 시장이름 유동인구테이블과 동일하게 변경

def change_name3(x):
    if x=='장위골목시장':
        return "장위전통시장"
    if x=='용문시장':
        return "용문전통시장"
    if x=='신원시장(신림1동)':
        return "신원시장"
    if x=='수유재래시장':
        return "수유전통시장"
    if x=='원당종합시장':
        return "인헌시장"
    if x=='마천중앙시장':
        return "마천시장"
    if x=='관악신사시장(신림4동)':
        return "관악신사시장"
    else: return x
    
df_price_address['시장이름'] = df_price_address['시장이름'].apply(change_name3)
```

```python
# 유동인구테이블에서 시장이름 가격-주소테이블테이블과 동일하게 변경

def change_name4(x):
    if x=='대림골목시장':
        return "대림시장"
    else: return x
    
df_customer['시장이름'] = df_customer['시장이름'].apply(change_name4)
```

```python
#가격-주소 테이블과 유동인구 테이블 merge
df = pd.merge(df_price_address, df_customer, on=['년도','분기','시장이름'], how='left')
df
```
